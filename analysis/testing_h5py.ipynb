```python
%matplotlib inline
import h5py

from matplotlib import pylab as pl
import numpy as np
```

```python
f = h5py.File('../results/predictNS/predictions_.hdf5','r')
```

```python
g = h5py.File('../data/hdf5datasets/NSMSDSRSCSTSRI.hdf5','r')
```

```python
g.items()
```

```python
preds = f.get('predictions')[:]
info = f.get('info')[:]
```

```python
info_orig = g.get('info')[:]
NS_orig = g.get('NETseq')[:]
```

```python
ix = 128
tff = (info_orig[:,0]==info[ix,0])& (info_orig[:,1]==info[ix,1]) & (info_orig[:,2]==info[ix,2])& (info_orig[:,3]==info[ix,3])

vec = NS_orig[tff,0,:].T
vec /=np.sum(vec)
pl.plot(vec);
pl.plot(preds[ix],'r');
pl.savefig('sample_'+str(ix)+'.pdf')
```
